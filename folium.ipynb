```python
import csv # csv 읽기,쓰기
import folium # 지도생성

# 파일 읽기 (utf-8, euc-kr)
file = open('csv/부산광역시_북구_도시공원정보_20210506_1620294900014_17874.csv',
        'r',encoding="euc-kr")
rdr = csv.reader(file)

# column명 skip
next(rdr,None) 

# 읽은 파일 내용보기(공원명[1], 위도[5], 경도[6])
for idx,tmp in enumerate(rdr):
    # 최초 1번 객체 만들기 시작. 주소의 중심이 location에 정의된 곳. (위도,경도,줌size,줌start)
    if idx == 0:
        map_data = folium.Map(location = [float(tmp[5]),float(tmp[6])],zoom_size=12, zoom_start=13)
    # mark표시 (위도,경도,이름)
    folium.Marker([float(tmp[5]),float(tmp[6])],popup=tmp[1]).add_to(map_data)
    print(idx,tmp[1],tmp[5],tmp[6]) # tmp[5],[6]는 str형 이므로 float형태로 변환해야 출력이 된다.
# 지도를 map폴더 html로 저장
map_data.save('map/ex01.html')
```

```python
import csv
import requests
import pandas as pd

# 데이터 1000개 => 800개 훈련, 200개 테스트 데이터
csv_url = "http://ihongss.com/resources/train.csv"

with requests.Session() as s:
    download = s.get(csv_url)
    decode_content = download.content.decode('utf-8')
    
    file = csv.reader(decode_content.splitlines(),delimiter=",")
    
    data = list(file)
    
    # 데이터 프레임 만들기(실제 데이터, column데이터)
    df = pd.DataFrame(data[1:],columns=data[0])
df
# df.info()
# df.describe()

#     for tmp in data:
#         print(tmp)
    
# df = pd.read_csv('http://ihongss.com/resources/train.csv')
# df
```
